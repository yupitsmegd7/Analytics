```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
r=pd.read_csv('/content/adult.csv')
```

```python
#2) Find Shape and Info of the Dataset
r.head()
r.info()
```

```python
#3) Fetch Random Sample From the Dataset
r.sample(frac=0.5,random_state=512)
```

```python
#4)Check Null Values In The Dataset
r.isnull().any()
```

```python
#5)Perform Data Cleaning [ Replace '?' with NaN ]
r.isin(['?']).sum()
import numpy as np
r['workclass']=r['workclass'].replace({'?':np.nan})
r['occupation']=r['occupation'].replace({'?':np.nan})
r['native-country']=r['native-country'].replace({'?':np.nan})
r
```

```python
#6)Drop all The Missing Values
r.isnull().sum()
r.dropna(inplace=True)
```

```python
#7) Check For Duplicate Data and Drop Them
r.duplicated().sum()
r.drop_duplicates(inplace=True)
```

```python
#8)Get Overall Statistics About The Dataframe
r.describe()
```

```python
#9)Drop The Columns education-num, capital-gain, and capital-loss
r.drop(['educational-num','capital-gain','capital-loss'],axis=1,inplace=True)
```

```python
#10)What Is The Distribution of Age Column?
r['age'].describe()
c= r['age']
plt.hist(c,edgecolor='y',rwidth=0.8,label='age')
plt.xlabel('ages')
plt.ylabel('population')
plt.legend()
```

```python
#11) Find Total Number of Persons Having Age Between 17 To 48
len(r[r['age'].between(17,48)])
```

```python
#12)What is The Distribution of Workclass Column?
r['workclass'].describe()
r['workclass'].unique()
r['workclass'].value_counts()
plt.hist(r['workclass'],edgecolor='r')
```

```python
#13)How Many Persons Having Bachelors and Masters Degree?
j=r['education'].str.contains('bachelor',case=False)
k=r['education'].str.contains('master',case=False)
len(r[j | k])
```

```python
#14)Replace Salary Values ['<=50','>50'] With 0 and 1
d=r[r['income'].isin(['<=50K'])]
r['income'].replace(['<=50K','>50K'],value=[0,1],inplace=True)
r
```

```python
#15) Which Workclass Getting The Highest Salary
r.groupby('workclass')['income'].mean().sort_values(ascending=False).head(1)
```

```python
#16)How Has Better Chance To Get Salary greater than 50K Male or Female?
o=r.groupby('gender')['income'].mean()
k=['Female','Male']
p= []
for i in o:
  p.append(i)
  plt.title('Above 50K chance')
  plt.bar(k,p,color=['r','b'])
```
