```python
#5) Total Number of App Titles Contain Astrology
len(s[s['App'].str.contains('Astrology',case=False)])
```

```python
#6)Find Average App Rating
s['Rating'].mean()
```

```python
#7) Find Total Number of Unique Category
s['Category'].nunique()
```

```python
#8)Which Category Getting The Highest Average Rating
s.groupby('Category')['Rating'].mean().sort_values(ascending=False)

#<>.groupby->a (dataframe)
#a.[<>]->j (series)
#mean works on series and gives a series and sort then also works on series
```

```python
#9)Find Total Number of App having 5 Star Rating
len(s[s['Rating']==5.0])
```

```python
#10) Find Average Value of Reviews
s['Reviews'].dtype
s[s['Reviews']=='3.0M']
s['Reviews']= s['Reviews'].replace('3.0M',3.0)
s['Reviews']= s['Reviews'].astype('float')
s['Reviews'].mean()
```

```python
#11)Find Total Number of Free and Paid Apps
print('Free',len(s[s['Type']=='Free']))
print('Paid',len(s[s['Type']!='Free']))
```

```python
#12) Which App Has Maximum Reviews?
s[s['Reviews']== s['Reviews'].max()]['App']
```

```python
#13) Display Top 5 Apps Having Highest Reviews
j=s['Reviews'].sort_values(ascending=False).head(5).index
s.iloc[j]['App']

#loc- col name
#iloc- index and without ()
```

```python
#14)Find Average Rating of Free and Paid Apps
s[s['Type']=='Free']['Rating'].mean()
```

```python
#15)Display Top  5 Apps Having Maximum Installs
l=s['Installs'].sort_values(ascending=False).head(5).index
s.iloc[l]['App']
```
