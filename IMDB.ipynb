```python
#19) Classify Movies Based on Ratings [Excellent, Good, and Average]
 #7.0 and up-> Excellent &  5.0-7-0-> Good and rest Average

def Rata(x):
  if x>=7.0:
    return 'Excellent'
  elif x<7.0 and x>=5.0:
    return 'Good'
  else:
    return "Average"

i['Rata']=i['Rating'].apply(Rata)
i[['Title','Rata']]
```

```python
#20)Find Unique Values From Genre
l=[]
k=i['Genre'].unique()
for j in k:
  l.append(j.split(","))

p=[]
for k in l:
  for a in k:
    if a not in p:
      p.append(a)
l=pd.DataFrame(p)
l
```

```python
#21)How Many Films of Each Genre Were Made?
l=[]
k=i['Genre']
for j in k:
  l.append(j.split(","))

p=[]
for k in l:
  for a in k:
    p.append(a)
Counter(p)
```
